```python
import pandas as pd
```

```python
df = pd.read_csv('players_20.csv')
#set index
df.set_index('short_name', inplace = True)

# Select Column
df = df[['long_name', 'age', 'dob', 'height_cm', 'weight_kg','nationality', 'club']]
```

```python
df
```

# Selecting with a single value

iloc[row_position, column_position]

```python
#get the height of L.Messi
df.iloc[0, 3]
```

```python
#get the weight of the cristiano Ronaldo
df.iloc[1, 4]
```

```python
#get all rows inside the 'height_cm' Column
df.iloc[:, 3]
```

```python
#get all columns that correspond to the index 'L. Messi'
df.iloc[0, :]
```

# Selecting with a list of value

```python
#get all data about Messi and Ronaldo

df.iloc[[0,1]]
```

```python
#get height of Messi and Ronaldo
df.iloc[[0,1], 3]
```

```python
#get the height and weight of L. Messi
df.iloc[0, [3,4]]
```

```python
df.iloc[[0,1], [3,4]]
```

# Selecting a Range of data with a slice

start:stop(Note that the start is included but the stop is excluded)

```python
#Slice column labels age/club
#players = ['Messi', 'Ronaldo']
players = [0,1]
df.iloc[players, 1:7]
```

```python
#Slice index labels(top1 to top10 player)
#Columns = ['age', 'dob','height_cm', 'weight_kg', 'nationality', 'club']

df.iloc[0:10, 1:7]
```

# Selecting with Condition

iloc cannot accept boolean series,but only a bolean list. We have to use the list() function to convert a series into a boolean list

```python
#One condition: Select player with height above 180 cm 
#Columns = ['age', 'dob','height_cm', 'weight_kg']

columns = [1,2,3,4,]
df.iloc[list(df['height_cm']>180), columns] 
```

```python
#Multiple condition: Select player with height above 180 cm of nationality Argentina

df.iloc[list((df['height_cm']>180) & (df['nationality']== 'Argentina'))]
```
