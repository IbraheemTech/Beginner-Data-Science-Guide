```python
import pandas as pd
```

```python
df = pd.read_csv('players_20.csv')

df.set_index('short_name', inplace = True)

df= df[['age', 'dob','height_cm', 'weight_kg', 'nationality', 'club']]
```

```python
df
```

# Sample()

```python
#extract 10 random elements from the 'nationality' column
df['nationality'].sample(10, random_state = 1)
```

```python
#extract a random 20% sample of the dataframe
df.sample(frac = 0.2)
```

```python
#upsample: Increases the sample rate -> frac>1 (Note: replace parameter has to be true for frac parameter > 1)

df.sample(frac = 2, replace = True)
```
