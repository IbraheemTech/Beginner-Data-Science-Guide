```python
import pandas as pd
```

```python
df = pd.read_csv('players_20.csv')
df.set_index('short_name', inplace = True)
df = df[['long_name','age','dob', 'height_cm','weight_kg','nationality','club']]
df
```

# Copy()

## deep = True

```python
# deep = True by default (Modifications to the data or indices of the copy)
df_copy() = df.copy()
```

```python
#update value in original dataframe
df.loc['L. Messi', 'height_cm'] = 180
```

```python
df
```

## deep = False

```python
# deep = false (Any changes to the data of the original will be reflected
# in the shallow copy and vice versa, means all the changes that we do in df will reflect to df_copy())
df_copy = df.copy(deep = False)
```

```python
#update value in original data 
df.loc['Cristiano Ronaldo', 'height_cm'] = 200
```

```python
#shallow copy vs original dataframe
df
```
