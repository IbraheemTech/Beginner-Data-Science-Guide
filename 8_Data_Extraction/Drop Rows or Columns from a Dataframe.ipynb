```python
import pandas as pd
```

```python
df = pd.read_csv('players_20.csv')

df.set_index('short_name', inplace = True)

df= df[['age', 'dob','height_cm', 'weight_kg', 'nationality', 'club']]
```

```python
df
```

# Drop Row

```python
#drop one row\
# o represent row and 1 is for column
# df.drop('L. Messi', axis = 0)

#Index Parameter is another option to drop row
df.drop(index = ['L. Messi'])
```

# Drop Column



```python
#drop one column
#axis parameter
# df.drop('dob', axis = 1)

#Column Parameter
df.drop(columns = ['dob'])
```

```python
#drop column by postion/index(drop last column)

df.drop(df.columns[[-1]], axis = 1)
```

```python
#drop two or more column and update the data (inplace = True)
df.drop(['dob', 'weight_kg'], axis = 1 , inplace = True)
```
