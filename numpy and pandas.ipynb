Python is good for automation, while Excel is not designed to automate task. For excel you can create a macro or VVA to simplify the tasks, but that is
the limit. However, python can go beyond that with its hundred of free libraries.

4. Cross-Platform Capability:

# Adding a new Column

## assign()

When to use assign method?

*Add multiple column in a single line of code.

*When you need to overwrite the values of an existing column(best practice)

It returns a new object(copy) with all the original columns in addition to the new ones.

```python
#Create random numbers for our new score columns
score1 = np.random.randint(1,100, size= 1000)
score2   = np.random.randint(1,100, size=1000)
```

```python
# Creating a series using the random numbers
series1 = pd.Series(score1, index=np.arange(0,1000))
series2 = pd.Series(score2, index=np.arange(0,1000))
```

```python
#Using assign() to multiple columns
df_exam.assign(score1=series1, score2=series2)
```

```python
df_exam
```

```python
# To update the values we have to overwrite the df_exam varribale
df_exam = df_exam.assign(score1=series1, score2=series2)
```

```python
# Show Dataframe
df_exam
```
