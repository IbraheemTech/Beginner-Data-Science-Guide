# Creating a Conditional Column from more than 2 choices: np.select()

```python
import numpy as np
```

```python
# Create an array based on multiple Price_tiers (+2 choices)
# Create 'Condition' and 'Values' varriable

conditions = [
    df_laptops['Price_euros'] >=3000,
    ((df_laptops['Price_euros'] >=2000) & (df_laptops['Price_euros']<=3000)),
    ((df_laptops['Price_euros']>=800) & (df_laptops['Price_euros']<= 2000)),
    df_laptops['Price_euros']<=800
]
values = ['Too Expensive', 'Expensive', 'Affordable', 'Cheap']
```

```python
np.select(conditions, values, default='Unknown').astype(object)
```

```python
# Set it to a new column
df_laptops['Price_tiers'] = np.select(conditions, values)
```

```python
df_laptops
```

## Note on default in np.select():
When using np.select() to assign values based on multiple conditions, it is important to specify the default parameter. By default, NumPy uses 0 (an integer) as the fallback value for entries that don't match any condition. If your choicelist contains strings (e.g., 'Expensive', 'Affordable'), this creates a conflict in data types, resulting in a TypeError. 
To ensure consistent data types across all outputs, a string default (e.g., 'Unknown') should be explicitly provided when the choices are also strings.
