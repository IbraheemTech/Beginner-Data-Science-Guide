# drop_duplicates():

drop_duplicates() is a Pandas DataFrame method used to remove duplicate rows from your data.

✅ Purpose:
To clean data by removing repeated or identical rows—either from the entire DataFrame or based on specific columns.

```python
#drop duplicates in 2 or more column
df_laptops.drop_duplicates(['Company'])[['Company', 'Price_euros']].value_counts('Company')
```

```python
# sort dataframe ascending by company and price
df_laptops = df_laptops.sort_values(['Company', 'Price_euros'])
```

```python
df_laptops.drop_duplicates(['Company'], keep = 'first')[['Company', 'Price_euros']]
```

```python
df_laptops.drop_duplicates(['Company'], keep = 'last', ignore_index = True)
```

# Question

* Find the biggest and smallest in laptops of each company
*  using sort() values and duplicated() method(keep 'first','last' and 'false')
  

```python
# sort dataframe ascending by company and inches
# Smallest first and biggest last
df_sort = df_laptops.sort_values(['Company', 'Inches'])
```

```python
df_sort.drop_duplicates(['Company'], keep = 'first')[['Company', 'Inches']]
```

```python
df_sort.drop_duplicates(['Company'], keep = 'last')[['Company', 'Inches']]
```

## The reason we use:


* df_sort.drop_duplicates(['Company'], keep='first')  # Smallest
* df_sort.drop_duplicates(['Company'], keep='last')   # Biggest

...is because we’ve already sorted by Inches, so now we can confidently take the first and last row per company to get the smallest and largest screen sizes.
