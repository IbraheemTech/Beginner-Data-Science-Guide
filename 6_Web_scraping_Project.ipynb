# Organize all data in a Dictionary

```python
# Create a dictionary with original name of league as a key, dictionary help us to properly manage the data
dict_countries = {
    'Spanish la liga': 'SP1', ' Spanish Segunda Division': 'SP2',
    'German Bundesliga': 'D1',
    'English Premier League': 'E0', 'English league 1': 'E2', 'English league 2': 'E3'   
}
```

```python
dict_countries['Spanish la liga']
```

```python
for league in dict_countries:
    print(dict_countries[league])
```

```python

dict_historical_data = {}

# leagues = ['E0','E2', 'E3',]
frames = []
#looping through leagues and append it into the list
for league in dict_countries:
    frames = []
    for season in range(15,25):
        df = pd.read_csv(root + str(season) + str(season+1) + '/' + dict_countries[league] + '.csv' )
        df.insert(1, 'season', season)
        frames.append(df)
    df_concat = pd.concat(frames)
    dict_historical_data[league] = df_concat
```

## Few question regarding this code:


## 🧠 Why frames = [] must be inside the outer for league in dict_countries loop:
✅ What this does:
For each league, it creates a new list of DataFrames, one for each season from 2015 to 2024.

Then it concatenates them and stores the combined DataFrame in dict_historical_data.

❌ If frames = [] were outside the for league loop:
All DataFrames for every league and every season would go into the same list.

So when df_concat = pd.concat(frames) runs, it would contain mixed data from all leagues.

You’d get incorrect results, and memory usage would explode because the same frames list keeps growing.
